```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train=pd.read_csv('train.csv')
```

```python
train
```

```python
train[train['Cabin'].isna()]
```

```python
df=train[['PassengerId', 'Survived', 'Pclass', 'Name', 'Sex', 'Age', 'SibSp',
       'Parch', 'Ticket', 'Fare', 'Embarked']]
```

```python
df=df.dropna(subset=['Embarked'])
```

```python
df[df['Age'].isna()]
```

```python
sb=df['Age'].quantile(q=0.55)
df['Age']=df['Age'].fillna(value=sb)
```

```python
plt.hist([df['Age'][df['Survived']==0], df['Age'][df['Survived']==1]], histtype='barstacked',label=['0','1']);
plt.legend();
```

```python
df['Embarked']=df['Embarked'].factorize()[0]
```

```python
plt.hist([df['Embarked'][df['Survived']==0], df['Embarked'][df['Survived']==1]], histtype='barstacked',label=['0','1']);
plt.legend();
```

```python
plt.scatter(df['Embarked'],df['Sex'].factorize()[0], c=df['Survived']);
```

```python
df['Sex']=df['Sex'].factorize()[0]
```

```python
df
```

```python
plt.scatter(df['Parch'],df['SibSp'], c=df['Survived']);
```

```python
xtrain=df[['Pclass','Sex','Age','SibSp','Parch']]
ytrain=df['Survived']
```

```python
xtest=pd.read_csv('test.csv')
yid=xtest['PassengerId']
```

```python
sb=xtest['Age'].quantile(q=0.56)
xtest['Age']=xtest['Age'].fillna(value=sb)
```

```python
xtest=xtest[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch']]
```

```python
xtest['Sex']=xtest['Sex'].factorize()[0]
xtest
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.coef_
```

```python
ypred=model.predict(xtest)
```

```python
a=pd.DataFrame((yid,ypred))
```

```python
a=a.T
```

```python
a.columns=['PassengerId','Survived']
```

```python
a.to_csv('ktoutcome.csv',index=False)
```
